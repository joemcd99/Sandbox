```python
import pdfplumber
import pandas as pd

def scrub_iso_bic(pdf_path, output_excel):
    all_data = []
    
    with pdfplumber.open('ISOBIC.pdf') as pdf:
        # Loop through pages (skipping the cover page if necessary)
        for i, page in enumerate(pdf.pages[1:]):  # Starts from Page 2 where data begins
            table = page.extract_table()
            if table:
                # On the first data page, capture the header
                if not all_data:
                    all_data.extend(table)
                else:
                    # On subsequent pages, skip the header row if it repeats
                    all_data.extend(table[1:])
            print(f"Scrubbed page {i + 2}")

    # Convert to DataFrame
    df = pd.DataFrame(all_data[1:], columns=all_data[0])

    # Basic Cleaning: Remove newlines that PDF extraction often adds to addresses
    df = df.applymap(lambda x: x.replace('\n', ' ') if isinstance(x, str) else x)

    # Save to Excel
    df.to_excel('isobic2026.xlsx', index=False)
    print(f"Successfully wrote data to {output_excel}")

# Usage
scrub_iso_bic("ISOBIC.pdf", "ISO_BIC_Directory.xlsx")
```
